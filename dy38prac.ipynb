```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv('international-airline-passengers.csv')
df.head()
```

**EDA**

```python
df = df.rename(columns={"International airline passengers: monthly totals in thousands. Jan 49 ? Dec 60":'passengers'})
```

```python
df.head()
```

```python
df.sort_values(by='passengers',na_position='first')
```

```python
df.drop(index=144, inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df['passengers']=df['passengers'].astype(int)
```

```python
df.passengers.dtype
```

```python
plt.figure(figsize=[10,6])
plt.plot(df['Month'],df['passengers'])
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.title('Airline Monthly data')
plt.show()
```

```python
decompose = seasonal_decompose(df['passengers'], model='multiplicative', period=12)
decompose.plot()
```

```python
trend = decompose.trend.plot()
```

```python
seasonl = decompose.seasonal.plot()
```

```python
resi = decompose.resid.plot()
```

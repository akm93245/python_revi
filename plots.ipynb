```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame({'time':[0,5,10,15], 'speed':[0,50,100,80]})
df
```

bar plot

```python
sns.barplot(x='speed', y='time'  , data= df)
```

point plot

```python
sns.pointplot(x='speed', y='time'  , data= df)
```

scatter plot

```python
sns.scatterplot(x='speed', y='time'  , data= df)
```

```python
sns.lineplot(x='speed', y='time'  , data= df)
```

```python
sns.countplot( y='time'  , data= df)
```

```python
sns.histplot(x='speed', y='time'  , data= df)
```

```python
sns.kdeplot(x='speed', y='time'  , data= df)
```

```python
sns.ecdfplot(x='speed', y='time'  , data= df)
```

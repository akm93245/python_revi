```python
copa = pd.read_excel('copa_group_a.xlsx')
copa
```

splitting the datasets in two

```python
copa_1 = copa.sample(frac=0.5, random_state=1)
copa_1
```

dropping the other part

```python
copa_2 = copa.drop(copa_1.index)
copa_2
```

joining the two parts

```python
mundo = pd.concat([copa_1,copa_2],ignore_index=False)
mundo
```

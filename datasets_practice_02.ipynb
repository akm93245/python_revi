# making a dataframe

### making a dataset of Copa America 

>Group A

```python
Group_A = pd.DataFrame({'Team':['Argentina','Canada', 'Chile','Peru'],'Matches played':[3,3,3,3] ,'Won':[3,1,0,0], 'Draw':[0,1,2,1], 'Lost':[0,1,1,2], 'Points':[9,4,2,1]})
Group_A
```

```python
Group_A.to_excel('copa_group_a.xlsx')
```

>Group B

```python
Group_B = pd.DataFrame({'Team':['Vanezuela','Equador', 'Mexico','Jamaica'],'Matches played':['3','3','3','3'] ,'Won':['3','1','1','0'], 'Draw':['0','1','1','0'], 'Lost':['0','1','1','3'], 'Points':['9','4','4','0']})
Group_B
```

> Group C

```python
Group_C = pd.DataFrame({'Team':['Uraguay','Panama', 'USA','Bolivia'],'Matches played':['3','3','3','3'] ,'Won':['3','2','1','0'], 'Draw':['0','0','0','0'], 'Lost':['0','1','2','3'], 'Points':['9','6','3','0']})
Group_C
```

>Group D

```python
Group_D = pd.DataFrame({'Team':['Colombia','Brazil', 'Costa Rica','Paraguay'],'Matches played':['3','3','3','3'] ,'Won':['2','1','1','0'], 'Draw':['1','2','1','0'], 'Lost':['0','0','1','3'], 'Points':['7','5','4','0']})
Group_D
```

`Combining these subset to a final set`

```python
copa_america =pd.concat([Group_A, Group_B, Group_C, Group_D], ignore_index=True)
copa_america
```
